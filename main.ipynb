```python


def get_highly_correlated_features(df, threshold=0.9):
    """
    Finds features with correlation greater than the given threshold.

    Args:
        df (pd.DataFrame): Dataset with only numeric features
        threshold (float): Correlation threshold (default 0.9)

    Returns:
        List[str]: List of features recommended for removal
    """
    corr_matrix = df.corr().abs()  # Absolute correlations
    upper_triangle = corr_matrix.where(
        np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
    )

    to_drop = [
        column for column in upper_triangle.columns 
        if any(upper_triangle[column] > threshold)
    ]

    return to_drop
```

```python
high_corr_features = get_highly_correlated_features(df.corr(), threshold=0.9)
print("Highly Correlated Features to Drop:", high_corr_features)
print("length",len(high_corr_features))
```

```python
## droping them

df_new = df.drop(high_corr_features,axis=1)
```
